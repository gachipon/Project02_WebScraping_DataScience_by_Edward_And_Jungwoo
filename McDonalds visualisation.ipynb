```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Import Data

mcdonalds_data = pd.read_csv('McDonalds_HK_Breakfast_Nutritional_Data.csv')
mcdonalds_data.rename(columns = {'Unnamed: 0':'Meal'}, inplace = True)
mcdonalds_data
```

```python
plt.barh(mcdonalds_data['Meal'], mcdonalds_data['Energy (kcal)'])
plt.xlabel('Energy(Calories)')
plt.ylabel('Meal')
plt.title('Energy Levels for Meals in McDonalds')
plt.show()
```

```python
plt.barh(mcdonalds_data['Meal'], mcdonalds_data['Protein (g)'])
plt.xlabel('Protein (g)')
plt.ylabel('Meal')
plt.title('Protein Contents for Meals in McDonalds')
plt.show()
```

```python
y_axis = np.arange(len(mcdonalds_data['Meal']))

plt.barh(y_axis -0.2, mcdonalds_data['Total Fat (g)'], height=0.2, label = 'Total Fat (g)')
plt.barh(y_axis, mcdonalds_data['Saturated Fat (g)'], height=0.2, label = 'Saturated Fat (g)')
plt.barh(y_axis +0.2, mcdonalds_data['Trans Fat (g)'], height=0.2, label = 'Trans Fat (g)')

plt.yticks(y_axis, mcdonalds_data['Meal'])
plt.legend()
plt.show()
```

```python
y_axis = np.arange(len(mcdonalds_data['Meal']))

plt.barh(y_axis -0.2, mcdonalds_data['Carbohydrates (g)'], height=0.4, label = 'Carbohydrates (g)')
plt.barh(y_axis +0.2, mcdonalds_data['Sugars (g)'], height=0.4, label = 'Sugars (g)')

plt.yticks(y_axis, mcdonalds_data['Meal'])
plt.legend()
plt.show()
```

```python
plt.barh( mcdonalds_data['Meal'], mcdonalds_data['Sodium (mg)'])
plt.xlabel('Sodium (mg)')
plt.ylabel('Meal')
plt.title('Sodium Contents for Meals in McDonalds')
plt.show()
```
